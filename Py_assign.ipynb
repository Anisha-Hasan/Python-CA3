```python
#Which player got the maximum premium on the base price?
df['Premium'] = df['SOLD PRICE'] - df['BASE PRICE']
max_premium_index = df['Premium'].idxmax()
max_premium_player = df.loc[max_premium_index, 'PLAYER NAME']
print(max_premium_player)
```

```python
#Which players got the maximum premium offering on their base price?
df['Premium'] = df['SOLD PRICE'] - df['BASE PRICE']
df_sorted = df.sort_values(by='Premium', ascending=False)
top_5_players = df_sorted.head(5)
print(top_5_players)
```

```python
#What is the average SOLD PRICE for each age category?
df.groupby('AGE')['SOLD PRICE'].mean()
```

```python
#Average SOLD PRICE for Different Playing Roles in Each Age Category?
soldprice_by_age_role =df.groupby(['AGE', 'PLAYING ROLE'])['SOLD PRICE'].mean().reset_index()
print(soldprice_by_age_role)
```
